**Q1. Display Title of The Movie Having Runtime >= 180 Minutes ?**

```python
data[data['Runtime (Minutes)']>=180][['Title','Runtime (Minutes)']].set_index('Title')
```

**Q2. Find Which Year There Was The Highest Voting**

```python
df_Highest_Votin = data.groupby('Year')['Votes'].sum().sort_values(ascending=False).reset_index()
df_Highest_Votin
```

```python
sns.barplot(x='Year',y='Votes',data=df_Highest_Votin)
plt.title("Votes By Year")
plt.show()
```

**Q3. In Which Year There Was The Highest Revenue?**

```python
df_Highest_Revenue = data.groupby('Year')['Revenue (Millions)'].sum().sort_values(ascending=False).reset_index()
df_Highest_Revenue
```

```python
sns.barplot(x='Year',y='Revenue (Millions)',data=df_Highest_Revenue)
plt.title("Revenue By Year")
plt.show()
```

**Q4. Find The Average Rating For Each Director**

```python
df_director = data.groupby('Director')['Rating'].mean().sort_values(ascending=False).reset_index()
df_director
```

**Q5.What is Top 10 Lengthy Movies Title**

```python
sns.barplot(x=lengthy['Runtime (Minutes)'],y=lengthy.index)
plt.title('Top 5 Lengtly Movies')
plt.show()
```

**Q12. Classify Movies Based on Ratings [Good,Better and Best]**

```python
def rating(rating):
    if rating>=7.0:
        return 'Excellent'
    elif rating>=6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
data['rating_category']=data['Rating'].apply(rating)
```
